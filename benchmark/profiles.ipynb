```python
import pandas as pd
data = pd.read_csv('./profile_tut_basic.csv', index_col=0)
data = data.T.apply(lambda x : x)
neworder = ['Final Typecheck', 'Includes Check', 'Length Check', 'Substitution Inlining' ]
data=data.reindex(columns=neworder)
data
```

```python
plot = data.plot(kind='bar', stacked=True, rot=30)
fig = plot.get_figure()
fig.set_size_inches(12, 5)
fig.set_dpi(150)
plot.grid(axis='y', linewidth=0.2)
plot.set_xlabel("Applied Optimization")
plot.set_ylabel("Runtime [ms]")
```
